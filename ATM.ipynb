```python
from random import randrange 

class ATM : 
    def __init__(self) -> None:
        # In data,  key --> user_pin value --> saving balance
        self.data = {
            1234: 12, 
            3453: 100, 
            9034: 34, 
            3452: 344
        }
        
        # starting the procedure 
        self.initiate()
        
    # to check whether the given user_pin available in the data     
    def checker(self, input_pin) : 
        if input_pin in self.data : 
            return True 
        else : 
            return False 
       
    # returns the money of the given pin    
    def getter(self, pin) : 
        if pin in self.data : 
            return self.data[pin] 
        else : 
            return -1 
        
    # changes the data value according to the need 
    def setter(self, pin, money) : 
        if self.data[pin] + money >= 0 : 
            self.data[pin] += money 
            return True 
        else :
             return False 
    
    
    
    def initiate(self) :
        print("Please, Enter the pin : ") 
        user_pin_input_1 = int(input()) 
        print("Please, Enter the pin again : ")
        user_pin_input_2 = int(input()) 
        
        # checking that both input are same or not 
        if user_pin_input_1 != user_pin_input_2 :
            print("ERROR : Both input doesn't Match") 
            return 
        
        # checking the user_pin available in the data 
        if not self.checker(user_pin_input_1) : 
            print("ERROR: The given user account not present.")
        
        self.userChoices(user_pin_input_1)
        
    # the choices given to the account holder or user 
    def userChoices(self, user_pin):
        while True :         
            print("Please, select one option.")
            print("\t1. Check Saving Balance") 
            print("\t2. Deposit funds") 
            print("\t3.Withdraw Funds") 
            print("\t4. Transfer Funds")

            user_chioce = input()

            if user_chioce == '1' : 
                self.savingBalance(user_pin)
            elif user_chioce == '2' : 
                self.depositFund(user_pin) 
            elif user_chioce == '3' : 
                self.withdrawFunds(user_pin) 
            elif user_chioce == '4' : 
                self.transferFunds(user_pin)
            else : 
                print("ERROR: Wrong choice")
                break 
        
        
    # choice 1 
    def savingBalance(self, user_pin): 
        print("Your Saving Balance =", self.getter(user_pin))
    
    # choice 2
    def depositFund(self, user_pin) : 
        print("Please, Enter the Money :")
        # the money is readed by the machine 
        money = float(input())
         
        check = self.setter(user_pin, money) 
        if check : 
            print("Deposit is Successful.") 
        else : 
            print("Deposit is not Successful.")
            print("Please, Try Again") 
        
        self.savingBalance(user_pin)
        
    # chioce 3 
    def withdrawFunds(self, user_pin):
        print("Enter the Money : ")
        money = float(input()) 
        
        if self.setter(user_pin, -money): 
            print("Withdraw is Successful")
        else : 
            print("ERROR : Insufficient Balance") 
        
        self.savingBalance(user_pin)
        
    # choice 4
    def transferFunds(self, user_pin): 
        checking_balance = randrange(1, 100) 
        print("Current checking balance =", checking_balance)
        
        print("Enter the amount to be transfer from checking to saving account:") 
        to_debit = float(input()) 
        
        if checking_balance <  to_debit : 
            print("ERROR: Insufficient Balance") 
            return 
        
        checking_balance -= to_debit 
        self.setter(user_pin, to_debit) 
        
        print("Funds are transfered successfully.")
        print("Checking Balance =", checking_balance)
        
        self.savingBalance(user_pin)     
       
ATM()     
```
