## Concept of normality of a variable 

![Alt text](https://i0.wp.com/www.projectguru.in/wp-content/uploads/2020/03/1-19.jpg?w%3D544%26ssl%3D1)

**to check the noramality of the variable** :
  1. calc. mean and standard deviation 
  2. calc. 
      - x1 = (mean + sd) 
      - x2 = (mean - sd)
  3. count the no. of values b/w x1 and x2 / no. of rows
      - criteria 1 : >= 0.683 
      - criteria 2 : >= 0.955
      - criteria 3 : >= 0.997
- **criteria 1** : len(mean-std > data > mean+std)/no_of_rows
- **criteria 2** : len(mean-2*std > data > mean+2*std)/no_of_rows
- **criteria 3** : len(mean-3*std > data > mean+3*std)/no_of_rows
