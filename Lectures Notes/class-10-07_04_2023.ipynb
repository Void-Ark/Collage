# Simple Linear Regression

**For Simple Linear Regression,, there is one independent vairable which is used to predict the dependent variable**
- prediction algorithm
- Two Variables : 
    1. independent variable (X)
        - it is the one which you are using to predict the dependent var. 
    2. dependent var. (Y)
        - which you want to predict 
    3. B0 (beta noot) : constant 
    4. B1 (beta one) : constant
- **FORMULA : Yi = B0 + B1 * Xi**
- steps :
    1. create 2 df. 
        - Y = dependent var. df.
        - X = independent var. df.
    2. to the x df. we need to add a constant 
        - X = statsmodel.api.add_constant(X)
    3. to create the regression modle 
        - model = statsmodel.formula.api.ols(Y, X).fit()
    4. to see the summery of the model 
        - model.summery()
    5. from model summery : 
        - B0 = (const, coff) 
        - B1 = (X, coff)

```python
import pandas as pd
from statsmodels.api import OLS
from statsmodels.api import add_constant

# Y = B0 + B1 * X

df = pd.read_csv("./Hotel.csv") 
X = df['PercentOccupancy']
Y = df['Revenueindollar1000']
X = add_constant(X) 
model = OLS(Y,X).fit()

print(model.summary())

# Y = B0 + B1 * X
print(-60.3747 + 8.2317*(44.5))
```

### how to find the accuracy of the model
1. create a new var. in pre_revanue which will store the predicted revenue for every case in the dataset 
    - there should be 12 predicted revenue
2. create a new var. error
    - error = |actual - predicted|
3. rel_error = error - actual
4. calc. the mean of these 12 values



```python
import pandas as pd
import numpy as np
from statsmodels.api import OLS
from statsmodels.api import add_constant

df = pd.read_csv("./Hotel.csv") 

X = df['PercentOccupancy']
Y = df['Revenueindollar1000']

model = OLS(Y,X).fit() 
#print(model.summary())

pre_Y = model.predict()

errror = (Y - pre_Y).abs()

rel_error = errror / Y 
 
print(rel_error.mean())
```
